```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
x_train.dtype
```

```python
# select 10 images and normalize and adjust dtype to 'float64'
images = x_train[0:10, :, :, :]/255.0
images.shape , images.dtype
```

```python
# create Filters and show them

filters = np.zeros((3,3,3,2))

for r in range(3):
  for c in range(3):
    if c==r:
      filters [r,c,:,0]=1

for r in range(3):
  for c in range(3):
   if c+r == 2:
      filters[r,c,:,1]=1

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Filter 0')
plt.imshow(filters [:,:,:,0])

plt.subplot(1,2,2)
plt.title('Filter 1')
plt.imshow(filters [:,:,:,1])
```

```python
# make convalved images
convalved_image = tf.nn.conv2d(images, filters,strides=1,padding='SAME')
```

```python
# show convalved images
for i in range(10):
    # Assuming convalved_image is already defined
    plt.figure(figsize=(12,8))
    # First subplot
    plt.subplot(1,3,1)  # 1 row, 3 columns, first subplot
    plt.title(f'Image {i}', loc='center')
    plt.imshow(images[i])


    plt.subplot(1,3,2)
    plt.imshow(convalved_image[i,:,:,0],cmap='gray')
    plt.title('Apply Filter 0')

    # Second subplot
    plt.subplot(1, 3, 3)  # 1 row, 3 columns, third subplot
    plt.title('Apply Filter 1')
    #plt.axis('off')  # Hide axes
    plt.imshow(convalved_image[i,:,:,1],cmap='gray')
    plt.tight_layout()  # adjusts the spacing between subplots in a figure to minimize overlap and ensure that all elements are clearly visible.
    plt.show()
```
